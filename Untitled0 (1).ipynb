```python
df.head()
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# now we can see missing values in the data
```

```python
df.head()
```

```python
norm = df['normalized-losses'].astype('float').mean()
```

```python
df['normalized-losses'].replace(np.nan,norm,inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
bor = df['bore'].astype('float').mean()
stor = df['stroke'].astype('float').mean()
```

```python
df['bore'].replace(np.nan,bor,inplace=True)
df['stroke'].replace(np.nan,stor,inplace=True)
```

```python
horse = df['horsepower'].astype('float').mean()
peak = df['peak-rpm'].astype('float').mean()
```

```python
df['horsepower'].replace(np.nan,horse,inplace=True)
df['peak-rpm'].replace(np.nan,peak,inplace=True)
```

```python
df.dropna(subset=['price'],axis=0,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['normalized-losses'] = df['normalized-losses'].astype('int')
df[['bore','stroke']] = df[['bore','stroke']].astype('float')
df['horsepower'] = df['horsepower'].astype('float')
df['price'] = df['price'].astype('float')
df['peak-rpm'] = df['peak-rpm'].astype('float')
```
